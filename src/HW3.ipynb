```python
import numpy as np

#create df1 with 10 rows and 5 columns
data1 = np.arange(50).reshape(10,5)
df1 = pd.DataFrame(data1, columns=['Mon', 'Tues', 'Wed', 'Thurs', 'Fri'])

# create df 2 with 5 rows and 4 columns
# Create a NumPy array with normally distributed data
# normal_data = np.random.normal(loc=0, scale=10, size=20).reshape(5,4).astype(int)  # Mean=0, StdDev=1, 10 samples

data2 = np.arange(20).reshape(5,4)
df2 = pd.DataFrame(data2, columns=['Thurs', 'Fri', 'Sat', 'Sun'])

print("\ndf1:")
print(df1)
print("\ndf2:")
print(df2)

#this will bring the two data frames together
#NaN will be seen where there is no overlapping data
df3 = pd.merge(df1, df2, how='outer')   
print("\ndf3:")
print(df3)
```

```python
# We see NaN values in the merged data frame. This is because the two data frames do not have the same columns and number of rows.
```
