```python
# fit_transform will fit the imputer to the data and transform it in one step
# This will fill the missing values in the DataFrame with the mean of the column
# Be careful to not do this on your test data, as it will leak information from the training data into the test data
# Here, we are filling the missing values in the column 'd' with the mean of the column
df1['d'] = imputer.fit_transform(df1[['d']])
```

```python
# lets try out how to merge two dataframes
# This is useful for combining two DataFrames based on a common key
df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "a", "b"], "data1": pd.Series(range(7), dtype="Int64")})
df2 = pd.DataFrame({"key": ["a", "b", "d"], "data2": pd.Series(range(3), dtype="Int64")})
```

```python
df2
```

```python
df1
```

```python
# on key is the common column in both DataFrames
# how is the type of join to be performed.  left, right, inner, outer
# left means all rows from df1 and only matching rows from df2
# right means all rows from df2 and only matching rows from df1
# inner means only matching rows from both DataFrames
# outer means all rows from both DataFrames, with NaN values for missing values
pd.merge(df1, df2, on='key', how='right')
```

```python
# Trying again with different dataframes
# This is useful for combining two DataFrames based on multiple keys
left = pd.DataFrame({"key1": ["foo", "foo", "bar"],"key2": ["one", "two", "one"],"lval": pd.Series([1, 2, 3], dtype='Int64')})
right = pd.DataFrame({"key1": ["foo", "foo", "bar", "bar"],"key2": ["one", "one", "one", "two"],"rval": pd.Series([4, 5, 6, 7], dtype='Int64')})
```

```python
left
```

```python
right
```

```python
# This is useful for combining two DataFrames based on multiple keys
# The merge method allows you to specify multiple keys to join on
# In this case, we are joining on the columns 'key1' and 'key2'
# The how parameter specifies the type of join to be performed
combo = pd.merge(left, right, on=['key1','key2'], how='outer')
```

```python
combo
```

```python
# Concatenating two DataFrames
# This is useful for combining two DataFrames along a specific axis
pd.concat([left, right], axis=0) # Concatenate along rows (axis=0)
```

```python
import pandas as pd
from sklearn.datasets import load_iris # load the iris dataset
from sklearn.model_selection import train_test_split # for splitting the dataset into training and test sets
```

```python
# datasets has a lot of datasets, including the iris dataset
# The iris dataset is a classic dataset for classification tasks
iris = load_iris()

# Convert the iris dataset to a pandas DataFrame
# This is useful for converting the dataset to a format that can be easily manipulated and analyzed
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target column
df_iris['species'] = iris.target
```

```python
df_iris.head(5)
```

```python
# Splitting the dataset into training and test sets
# This is useful for splitting the dataset into training and test sets for model evaluation
df_train, df_test = train_test_split(df_iris, test_size=0.3, random_state=42)
```

```python
df_train.head(5)
```

```python
# Scaling the features
# This is useful for scaling the features to a specific range or distribution
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
scaler = StandardScaler()  # Use MinMaxScaler() for normalization between 0 and 1
df_train_scaled = scaler.fit_transform(df_train) # Fit the scaler to the training data and transform it

# Apply the same transformation to the test data
df_test_scaled = scaler.transform(df_test) # Use transform() instead of fit_transform() to avoid data leakage
# This will scale the test data using the same parameters as the training data
```

```python
df_test_scaled
# After scaling the data is in a numpy array format, so we need to convert it back to a DataFrame
```

```python
# Convert the scaled data back to a DataFrame
df_train_scaled = pd.DataFrame(df_train_scaled, columns=df_train.columns)
df_test_scaled = pd.DataFrame(df_test_scaled, columns=df_test.columns) 
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.hist(df_iris['petal length (cm)'], bins=50, color="black", alpha=0.3);
ax2.boxplot(df_iris['petal length (cm)']);
ax3.plot(df_iris['petal length (cm)'], color="orange", alpha=0.8);
ax4.scatter(df_iris['petal length (cm)'], df_iris['petal width (cm)']);
```

```python
# Another way to create subplots using matplotlib
# This is useful for creating multiple subplots in a single figure

fig2, axes = plt.subplots(2, 2, figsize=(5, 10)) # Create a 3x1 grid of subplots
fig2.set_size_inches(10, 5) # Set the size of the figure to 10x10 inches
# You now access plots using the axes array
# The first subplot is a histogram of the petal length
axes[0,0].boxplot([df_iris['petal length (cm)'], df_iris['petal width (cm)']]);
axes[0,0].set_title("This is a Title")
axes[0,0].set_xlabel("X-axis Label")
axes[0,0].set_ylabel("Y-axis Label")
# The second subplot is a line plot of the cumulative sum of the petal length
axes[1,1].plot(df_iris['petal length (cm)'].cumsum(), color="orange", linestyle="dashed", marker="o");
# The third subplot is a line plot of the petal length and petal width
axes[1,0].plot(df_iris['petal length (cm)'], color="red", linestyle="dashed", label="Default");
axes[1,0].plot(df_iris['petal width (cm)'], color="green", linestyle="dashed", drawstyle="steps-post", label="steps-post");
axes[1,0].set_title("This is a Title")
axes[1,0].set_xlabel("X-axis Label")
axes[1,0].set_ylabel("Y-axis Label")

axes[1,0].legend(loc="upper left")
fig2.subplots_adjust(wspace=2, hspace=1)
```
